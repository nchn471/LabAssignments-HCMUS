## Cài đặt lại thuật toán `Vertical Apriori`

### Xử lý dữ liệu

Chuyển `DataFrame` về dạng `List`

```python
records = []
for i in range(0,dataframe.shape[0]):
    records.append([str(dataframe.values[i,j]) for j in range (0, dataframe.shape[1])])
```

Xử lý chuỗi `'nan'`

```python
records = [[item for item in i if item != 'nan'] for i in records]
records
```

Chuyển database sạng dạng `vertical`

```python
def convert_vertical(records):
    transactions = {}
    for transaction_index, transaction in enumerate(records):
        for t in transaction:
            # frozenset là tập không thay đổi (immutable set) tức nó không cho phép chỉnh sửa, thay đổi hay giá trị lặp
            t = frozenset([t])
            if t not in transactions:
                transactions[t] = []
            transactions[t].append(transaction_index)
    return transactions
```

```python
X = convert_vertical(records)
for i in X:
    print(i,"\t:",X[i])
```

### Cài đặt thuật toán

Cơ sở lý thuyết

- $C_k$: k-item candidate itemset (tâp ứng viên có k-hạng mục)

- $F_k$: k-item frequent itemset (được lọc ra từ $C_k$ với các giá trị thoã mãn `minimum_support`)

Tập $C_1$ được tạo riêng làm cơ sở để áp dụng các hàm sau

```python
# Vì C_1 chỉ có một item duy nhất nên ta chỉ cần lấy các key của database X
# Vì X là database dạng vertical nên các key của nó chính là các itemsets (tập hạng mục)
def create_candidate_1_itemsets(X):
    c1 = []
    for transaction in X:
        c1.append(transaction)
    return c1
```

Tính $F_k$ (tập hạng mục thường xuyên)

```python
# Hàm trả về tổng số giao dịch (với dữ liệu ngang thì nó là số dòng)
# Vì đây là dữ liệu dọc nên nó sẽ là số giao dịch lớn nhất trong danh sách tid
def find_num_transactions(X):

    num_transactions = 0
    for item in X:
        for tid in X[item]:
            if tid > num_transactions:
                num_transactions = tid 

    # Trả về tổng số giao dịch + 1 (vì Python bắt đầu đếm từ 0)
    return num_transactions + 1

def create_frequent_itemsets(X, ck, min_support):
    # Danh sách tập hợp mục thường xuyên
    freq_item = []
    # Tạo từ điển để lưu trữ hỗ trợ của mỗi tập hợp mục
    item_support = {}
    # Tìm tổng số giao dịch trong tập dữ liệu
    num_transactions = find_num_transactions(X)

    # Duyệt qua từng tập hạng mục ứng viên (tập ck)
    for items in ck:
        # tid_list lưu trữ các tid chung của các tập hạng mục
        tid_list = None  
        # Duyệt qua mỗi mục trong tập hạng mục
        for item in items:
            # Chuyển về dạng frozenset đảm bảo tính duy nhất và độc lập
            item = frozenset([item])
            # Nếu tid_list = None (tức là hạng mục đầu tiên)
            if tid_list is None:
                # Khởi tạo set với các hạng mục đầu tiên
                tid_list = set(X[item])  
            else:
                # Thực hiện phép giao lấy tid chung của các hạng mục trong tập hạng mục
                tid_list = tid_list.intersection(X[item]) 
        # Tính support = số tid chung / tổng giao dịch
        item_support[items] = len(tid_list) / num_transactions
        
    # Duyệt qua từng tập hạng mục và kiểm tra support
    for item in item_support:
        # Nếu hỗ trợ lớn hơn hoặc bằng min_support, thêm vào danh sách tập hạng mục thường xuyên
        if item_support[item] >= min_support:
            freq_item.append(item)
            
    # Trả về danh sách tập hạng mục thường xuyên và từ điển hỗ trợ
    return freq_item, item_support
```

Tính $C_k$ từ $F_{k-1}$

```python
# Hàm hỗ trợ tính tổ hợp từ tập hợp các phần tử cho trước
from itertools import combinations

# Hàm tạo k-item candidate itemsets
def create_candidate_k_itemsets(freq_item, k):
    ck = []
    
    # Với k = 1, tạo 2-item candidate itemsets (c2)
    if k == 1:
        for f1, f2 in combinations(freq_item, 2):
            item = f1 | f2 # union of two sets
            ck.append(item)
    # Với k != 1
    else:    
        for f1, f2 in combinations(freq_item, 2):

            # Nếu k-itemset có k - 1 phần tử chung 
            # sẽ lấy phần hợp của chúng vào k+1-itemset
            intersection = f1 & f2
            if len(intersection) == k - 1:
                item = f1 | f2
                if item not in ck:
                    ck.append(item)
    return ck
```

Thuật toán `vertical apriori` chính trong đó:


- `X` là tập dữ liệu chứa các giao dịch (vertical). 
- `min_support` là ngưỡng hỗ trợ tối thiểu để một tập hạng mục được coi là thường xuyên.
- `min_combination` chỉ trả về các giao dịch có ít nhất **min_combination items** (mặc định là 1)
- `max_combination` chỉ trả về các giao dịch có tối đa **max_combination items** (mặc định bằng tất cả số hạng mục mà database có).

```python
def apriori_from_scratch(X, min_support, min_combination=1, max_combination=None):

    # Khởi tạo max_combination nếu không được cung cấp
    if max_combination is None:
        max_combination = len(X)

    # Tạo tập hợp ứng viên 1-item (c1)
    c1 = create_candidate_1_itemsets(X)
    # Tạo tập hợp itemsets thường xuyên f1 từ c1
    freq_item, item_support_dict = create_frequent_itemsets(X, c1, min_support)
    # Thêm f1 vào freq_items (tập hợp các itemsets thường xuyên)
    freq_items = [freq_item]

    # Bắt đầu lặp từ k = 1
    k = 1
    while True:
        # Lấy tập hợp itemsets thường xuyên fk từ freq_items (được lưu dưới dạng list, bắt đầu từ index 0)
        freq_item = freq_items[k-1]
        # Tính tập hợp ứng viên ck+1 từ fk
        ck = create_candidate_k_itemsets(freq_item, k)
        # Tính tập hợp itemsets thường xuyên fk+1 từ ck+1 và từ điển hỗ trợ
        freq_item, item_support = create_frequent_itemsets(X, ck, min_support)
        # Nếu fk+1 trống => dừng
        # Kiểm tra điều kiện dừng để tránh thêm list rỗng vào freq_items
        if len(freq_item) == 0:
            break

        # Thêm fk+1 vào freq_items 
        freq_items.append(freq_item)
        # Cập nhật từ điển hỗ trợ tổng
        item_support_dict.update(item_support)
        # Tăng k để tiếp tục lặp
        k += 1

    # Chọn các tập hợp itemsets thường xuyên trong khoảng min_combination đến max_combination
    freq_items = freq_items[min_combination-1:max_combination-1]
        
    return freq_items, item_support_dict
```

### Kiểm tra và so sánh kết quả

```python
X = convert_vertical(records)
min_support = 0.5
min_combination = 2
max_combination = len(X)
```

```python
freq_items, item_support_dict = apriori_from_scratch(X,min_support, min_combination, max_combination)
freq_items
```

Để dễ nhìn hơn ta làm sạch nó

```python
def tidy(freq_items, item_support_dict):
    supports = []
    itemsets = []
    for itemset in freq_items:
        for item in itemset:            
            supports.append(item_support_dict[item])
            itemsets.append(item)
    df = pd.DataFrame({'Itemset': itemsets, 'Support': supports})
    df_sorted = df.sort_values(by=['Support'], ascending=False)
    return df_sorted

tidy(freq_items, item_support_dict)
```
