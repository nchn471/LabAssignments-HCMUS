#Data Preprocessing

```python
data = df.copy()
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for col in cat_cols:
    data.loc[:, col] = encoder.fit_transform(data[col])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.values)
```

```python
scaled_data.shape
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
elbow_method = KElbowVisualizer(KMeans(), k=(1,11))
```

```python
elbow_method.fit(scaled_data)
elbow_method.show()
```

```python
kmeans_model = KMeans(n_clusters=3)
kmeans_model.fit(scaled_data)
```

```python
# xut ra label ca tp training (scaled_data) ta dng .labels_
kmeans_model.labels_
```

```python
# d on ra label ca tp testing (ta gi s l scaled_data) ta dng .predict
kmeans_model.predict(scaled_data)
```

```python
# xut ra cc centroids ta c th dng .cluster_centers_
kmeans_model.cluster_centers_.shape
```

##PCA

```python
from sklearn.decomposition import PCA #Khai bo th vin PCA
pca = PCA(n_components=len(data.columns)) #khai bo
pca.fit(scaled_data)
```

```python
#cho ta bit c cc thnh phn ph c bao nhiu lng thng tin ca data gc
pca.explained_variance_ratio_*100
```

##3 dimensional PCA

```python
pca = PCA(n_components=3) #dng pca 3 thnh phn
pca_df = pd.DataFrame(pca.fit_transform(scaled_data), columns=["pca_1", "pca_2", "pca_3"])#to ra dataframe vi 3 thnh phn l pca1, pca2, v pca3
```

```python
pca_df
```

```python
plt.figure(figsize=(12, 7))
ax = plt.subplot(projection="3d")
ax.scatter(pca_df.pca_1, pca_df.pca_2, pca_df.pca_3, c=kmeans_model.labels_, cmap="cool")
plt.show()
```

##2 dimensional PCA

```python
pca = PCA(n_components=2)
pca.fit(scaled_data)
pca_df = pd.DataFrame(pca.transform(scaled_data), columns=["pca1", "pca2"])
```

```python
pca_df.head()
```

```python
plt.figure(figsize=(12, 7))
ax = plt.subplot()
ax.scatter(pca_df.pca1, pca_df.pca2, c=kmeans_model.labels_, cmap="cool")
plt.show()
```

##Nu mun v c thm legend

```python
plt.figure(figsize=(12, 7))
ax = plt.subplot()
group = kmeans_model.labels_
for g in np.unique(group):
    index = np.where(group==g)
    ax.scatter(pca_df.iloc[index].pca1, pca_df.iloc[index].pca2, label=g, cmap="cool")
ax.legend()
plt.show()
```
